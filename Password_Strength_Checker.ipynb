```python
import time
```

```python
'''Password Checker'''

password =(input('Enter a password: '))

if len(password) < 8:
  print('Password is not up to 8 characters')
else:
  print("Password has passed the length test ✅")

time.sleep(1)
if password.isalpha():
  print("Print password has passed the alphabetic test ✅")
  time.sleep(1)
else:
  print("Password has failed the alphabetic test ❌")
  decision1 = input("Your password has no alphabets, do you wish to continue regardless? ")
  if decision1 == 'Yes':
    print(f'Your password is: {password},  keep it safe')
  if decision1 == 'No':
    new_pass = input("Enter a new password that contains alphabets: ")
    print(f'Your new password is: {new_pass}')
```
